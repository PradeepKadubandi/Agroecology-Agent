```python
import gym
from gym.envs.registration import register
from baselines import deepq
import environment
```

```python
env_id = 'AgroEcology-v0'
register(
      id=env_id,
      entry_point='environment:EnvironmentV2',
  )
```

```python
env = gym.make(env_id, field_size=10)
```

```python
print ("field_size = {}".format(env.field_size))
env.render('human')
```

```python
act = deepq.learn(
    env,
    network='mlp',
    lr=1e-3,
    total_timesteps=100000,
    buffer_size=5000,
    exploration_fraction=0.1,
    exploration_final_eps=0.02,
    print_freq=25
)
```

```python
for i in range(5):
    obs, done = env.reset(), False
    episode_rew = 0
    while not done:
        obs, rew, done, _ = env.step(act(obs[None])[0])
        episode_rew += rew
    print ("Environment at end of episode:")
    env.render('human')
    print("Episode reward", episode_rew)
```
