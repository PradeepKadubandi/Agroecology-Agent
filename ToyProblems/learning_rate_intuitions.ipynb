```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
n = 100
alpha = 0.001
coeffs = [alpha * math.pow(1-alpha, t) for t in range(n)]
coeffs = coeffs + [math.pow(1-alpha, n), ]
print (coeffs)
print (sum(coeffs))
```

```python
N = 500
t = np.array([i for i in range(1, N+1)])

one_over_t = 1 / t

start_alpha = 0.01
decay_iteration = 100
decay_rate = 0.9
decaying_alpha = np.array([start_alpha * math.pow(decay_rate, i / decay_iteration) for i in range(N)])

for i in range(N):
    if decaying_alpha[i] > one_over_t[i]:
        first_index = i
        break
        
print ('First index where decay alpha is higher than 1/n : {}'.format(first_index))

plt.figure()
plt.plot(t, one_over_t, t, decaying_alpha)
plt.show()
```
